```python
# do similar operations for all columns that contain non-numeric data
data = pd.concat([data.drop("occupation", axis=1), pd.get_dummies(data.occupation).add_prefix('occupation_')], axis=1)
data = pd.concat([data.drop("workclass", axis=1), pd.get_dummies(data.workclass).add_prefix('workclass_')], axis=1)

data = pd.concat([data.drop("education", axis=1)])

data = pd.concat([data.drop("marital-status", axis=1), pd.get_dummies(data['marital-status']).add_prefix('marital-status_')], axis=1)
data = pd.concat([data.drop("relationship", axis=1), pd.get_dummies(data.relationship).add_prefix('relationship_')], axis=1)
data = pd.concat([data.drop("race", axis=1), pd.get_dummies(data.race).add_prefix('race_')], axis=1)
data = pd.concat([data.drop("native-country", axis=1), pd.get_dummies(data['native-country']).add_prefix('native-country_')], axis=1)
```

```python
# updated data output
data
```

```python
# in the 'gender' column, replace all 'Male' values ​​with 1 and all 'Female' values with 0
data['gender'] = data['gender'].apply(lambda x: 1 if x == 'Male' else 0)
# a similar operation with the 'income' column - if a person's income is more than 50k, then it takes the value 1, if less, then 0
data['income'] = data['income'].apply(lambda x: 1 if x == '>50K' else 0)
```

```python
# output correlation between features
data.corr()

# Correlation measures the degree of linear relationship between two variables.
# The correlation value ranges from -1 to 1, where:

#1: positive linear correlation (both traits increase together)
# -1: negative linear correlation (one trait increases, the other decreases)
#0: no linear correlation

# In this case, the correlation between features looks like this:

# the values ​​at the intersection of a row and a column represent the correlation coefficients between the corresponding variables.
# For example, the value 0.088120 in the cell where the strings "age" and "gender" intersect,
# denotes the correlation coefficient between age and gender.

# Interpretation of correlation coefficients:

# for example age and gender (age and gender): 0.088120.
# This is a positive coefficient, which may indicate some positive relationship between age and gender.
# However, the value is relatively small, so the connection may be weak.

#fnlwgt and capital-gain: -0.003706.
# This value close to zero indicates a weak negative correlation between fnlwgt and capital-gain.

# educational-num and capital-gain: 0.125146. This is a positive coefficient
# which may indicate a stronger positive relationship between educational attainment and capital gains.

# identical features will always correlate with each other with the result "1"
```

```python
plt.figure(figsize=(18, 12))
sns.heatmap(data.corr(), annot=False, cmap='coolwarm')

# This line of code uses the Seaborn library to create a heatmap
# correlations between numerical features in DataFrame data

# data.corr(): Calculates the correlation coefficient between all numeric features in the DataFrame data.
# The result is a new DataFrame containing correlation values ​​between feature pairs.

# sns.heatmap(...): Uses the heatmap function from the Seaborn library to visualize the correlation matrix.
# The options in parentheses configure the appearance of the heatmap:

# cmap='coolwarm': Specifies a color map to display correlation values.
```

```python
# variable calculates absolute values ​​of coefficients
# correlations between data columns and the target variable (income).
correlations = data.corr()['income'].abs()
# Sorts correlation coefficients in ascending order.
sorted_correlations = correlations.sort_values()
# Specifies the number of columns (features) to remove.
# This number is equal to 80% of the total number of data columns.
num_cols_to_drop = int(0.8 * len(data.columns))
# Selects the first num_cols_to_drop columns from the sorted list,
# which have the smallest absolute correlation coefficients with the target variable.
cols_to_drop = sorted_correlations.iloc[:num_cols_to_drop].index
# Directly removes columns from the data.
data_updated = data.drop(cols_to_drop, axis = 1)
```

```python

# as a classifier we exclude the RandomForestClassifier method
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split    


# using this principle you can reset any parameter
data = data.drop('fnlwgt', axis=1)

train_data, test_data = train_test_split(data, test_size=0.2)
```

```python
# train data output
train_data
```

```python
# test data output
test_data
```

```python
# section of training data for features
train_x = train_data.drop("income", axis = 1)
# and labels (what we will predict)
train_y = train_data["income"]
```

```python
# the same for test data, for features
test_x = test_data.drop("income", axis = 1)
# and for tags
test_y = test_data["income"]
```

```python
# RandomForestClassifier method applying
model = RandomForestClassifier()

model.fit(train_x, train_y)
```

```python
# performance of the RandomForestClassifier method on test data
model.score(test_x, test_y)
```

```python
# output numeric values ​​for all features (afterwards we place them in the values ​​variable)
values = model.feature_importances_
```

```python
# display the names of the features (then put them in the names variable)
names = model.feature_names_in_
```

```python

# combine the feature names and their numeric values ​​into one dictionary using the zip function

importances = dict(zip(names, values))
```

```python
# this code actually creates a new importances dictionary,
# in which the keys are stored and the values ​​are sorted in descending order.

importances = {k: v for k, v in sorted(importances.items(), key = lambda x : x[1], reverse = True) }

# importances.items() returns a representation of the dictionary as a sequence of (key, value) tuples.
# key=lambda x: x[1] specifies a function to extract the value (the second element) from each tuple. This is used to sort by values.

# reverse=True indicates sorting in descending order.

# {k: v for k, v in ...}:
# This is a dictionary generator that creates a new dictionary,
# traversing a sorted sequence of tuples.
#k:v specifies the key and value for each pair in the new dictionary.

```

```python
# display the updated dictionary starting with the most significant variable for prediction (it will be the first).
# Next, those that have less importance for the final prediction will be displayed
importances
```

```python
# use GridSearchCV to tune hyperparameters to get more
# optimal result, accuracy at this point (model performance - 0.8512642030914116)
# to check uncomment the code

# model.score(test_x, test_y)

from sklearn.model_selection import GridSearchCV

# dictionary containing parameters for hyperparameters, where n_estimators is the number of trees,
# in this case
# this is a list - [100, 200, 300], this means that this is the number of trees,
# which the model will iterate over to identify
# most accurate results

# max_depth': [5, 10, 15, None] - random tree depth
# None means that the tree will be built until
# until the nodes reach the minimum number of samples to split
# (min_samples_split)
# here the model will also iterate over all options from [5, 10, 15, None]
# to identify the most accurate

# min_samples_split - minimum number of samples for a section

# max_features - maximum number of features to split


# When verbose is set to True or some integer value,
# The function will provide more detailed output about the cross-validation and grid search process. 

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [5, 10, 15, None],
    'min_samples_split': [2, 4],
    'max_features': ['sqrt', 'log2']
    }

grid_search = GridSearchCV(estimator=RandomForestClassifier(),
                           param_grid=param_grid, 
                           verbose=10)

```

```python

# testing the model with updated parameters
grid_search.fit(test_x, test_y)
```

```python
# this code will provide the best parameters for the model based on tests using GridSearchCV
grid_search.best_estimator_
```

```python
# pass the resulting best hyperparameters to the model variable
model = grid_search.best_estimator_
```

```python
# check the accuracy of the model with new parameters
# as you can see, the result increased from 0.85... to 0.895 (almost 0.9). While the maximum value is 1,
# the result is very good
model.score(test_x, test_y)
```
